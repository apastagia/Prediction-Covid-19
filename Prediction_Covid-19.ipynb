```python
india_cases = covid[covid['location'] == 'India']
```

```python
india_cases.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Total cases per day
sns.set(rc={'figure.figsize': (15, 10)})
sns.boxplot(x = 'date', y = 'total_cases', data = india_cases)
plt.show()
```

```python
##plot few records on graph
india_last_10_days = india_cases.tail(10)
```

```python
sns.set(rc={'figure.figsize': (15, 10)})
sns.lineplot(x = 'date', y = 'total_cases', data = india_last_10_days)
plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 10)})
sns.boxplot(x = 'date', y = 'total_tests', data = india_cases)
plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 10)})
sns.lineplot(x = 'date', y = 'total_tests', data = india_last_10_days)
plt.show()
```
